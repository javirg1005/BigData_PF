Creación de KPIs

```python
import json
```

Mayor valor de mercado de jugadores

```python
df_mv_players = spark.sql("SELECT Player as `Jugadores`, CAST(REPLACE(REPLACE(FMV.Market_value, '€', ''), 'm', '') AS FLOAT) as `Valor de mercado (millones €)` FROM FMV ORDER BY `Valor de mercado (millones €)` DESC LIMIT 20;")
df_mv_players.show()
df_mv_players = df_mv_players.toPandas()
mv_players = df_mv_players.to_json()
mv_players = json.loads(mv_players)
mv_players

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_players.json', 'w') as file:
    json.dump(mv_players, file, indent=4)
```

Mayor valor de mercado de clubs

```python
df_mv_clubs = spark.sql("SELECT Club_x as `Equipos`, SUM(CAST(REPLACE(REPLACE(FMV.Market_value, '€', ''), 'm', '') AS FLOAT)) as `Valor de mercado (millones €)` FROM FMV WHERE Club_x != 'null' GROUP BY Club_x ORDER BY `Valor de mercado (millones €)` DESC LIMIT 20;")
df_mv_clubs.show()
df_mv_clubs = df_mv_clubs.toPandas()
mv_clubs = df_mv_clubs.to_json()
mv_clubs = json.loads(mv_clubs)
mv_clubs

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_clubs.json', 'w') as file:
    json.dump(mv_clubs, file, indent=4)
```

Jugadores con más tarjetas

```python
df_cards = spark.sql("SELECT PLAYERS.pretty_name as `Jugador`, SUM(red_cards) as `Tarjetas Rojas`, SUM(yellow_cards) as `Tarjetas Amarillas` FROM APPEARANCES JOIN PLAYERS ON APPEARANCES.player_id = PLAYERS.player_id GROUP BY `Jugador` ORDER BY `Tarjetas Rojas` DESC LIMIT 20")
df_cards.show()
df_cards = df_cards.toPandas()
cards = df_cards.to_json()
cards = json.loads(cards)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_cards.json', 'w') as file:
    json.dump(cards, file, indent=4)
```

País con más ligas

```python
df_competitions = spark.sql("SELECT country_name as `País`, count(name) as `Nº de ligas` FROM COMPETITIONS WHERE country_name != 'null' GROUP BY country_name")
df_competitions.show()
df_competitions = df_competitions.toPandas()
leagues = df_competitions.to_json()
leagues = json.loads(leagues)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_leagues.json', 'w') as file:
    json.dump(leagues, file, indent=4)
```

País con mas jugadores

```python
df_cp = spark.sql("SELECT country_of_birth as `País`, count(pretty_name) as `Nº de jugadores` FROM PLAYERS WHERE country_of_birth != 'null' GROUP BY country_of_birth LIMIT 30")
df_cp.show(50)
df_cp = df_cp.toPandas()
cp = df_cp.to_json()
cp = json.loads(cp)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_cp.json', 'w') as file:
    json.dump(cp, file, indent=4)
```

Goles totales por equipo

```python
df_gt = spark.sql("SELECT Club_x as `Equipos`, SUM(Gls) as `Goles totales` FROM FMV GROUP BY `Equipos` ORDER BY `Goles totales` DESC LIMIT 25")
df_gt.show(50)
df_gt = df_gt.toPandas()
gt = df_gt.to_json()
gt = json.loads(gt)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_gt.json', 'w') as file:
    json.dump(gt, file, indent=4)
```

Goles totales por jugadores (ranking 20)

```python
df_gavgp = spark.sql("SELECT PLAYERS.pretty_name as `Jugador`, SUM(APPEARANCES.goals) as `Goles totales` FROM PLAYERS JOIN APPEARANCES ON PLAYERS.player_id = APPEARANCES.player_id GROUP BY `Jugador` ORDER BY `Goles totales` DESC LIMIT 20")
df_gavgp.show()
df_gavgp = df_gavgp.toPandas()
gavgp = df_gavgp.to_json()
gavgp = json.loads(gavgp)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_gavgp.json', 'w') as file:
    json.dump(gavgp, file, indent=4)
```

Valor de mercado total de los países

```python
df_mvpais = spark.sql("SELECT Nation as `País`, SUM(CAST(REPLACE(REPLACE(Market_value, '€', ''), 'm', '') AS FLOAT)) as `Valor de mercado (millones €)` FROM FMV WHERE NAtion != 'null' GROUP BY `País` ORDER BY `Valor de mercado (millones €)` DESC")
df_mvpais.show(50)
df_mvpais = df_mvpais.toPandas()
mvpais = df_mvpais.to_json()
mvpais = json.loads(mvpais)

with open('/content/drive/MyDrive/DatosCollabBigData/PF/mv_mvpais.json', 'w') as file:
    json.dump(mvpais, file, indent=4)
```
